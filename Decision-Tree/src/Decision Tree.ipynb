```python
import pandas as pd
import numpy as np
from sklearn import tree
```

```python
def read_csv(file):
    df = pd.read_csv(file)
    df.columns = df.columns.str.replace('\W','')
    df['Occupied'] = df['Occupied'].str.replace('[^[a-zA-Z]', '')
    df['Enjoy'] = df['Enjoy'].str.replace('[^[a-zA-Z]', '')
    return df
```

```python
df = read_csv("../data/dt-data.csv")
df.head()
#df.dtypes
```

```python
def encoding(df):
    obj_df = df.select_dtypes(include=['object'])
    obj_df = pd.get_dummies(obj_df, columns=list(obj_df.columns)[:-1])
    #obj_df = obj_df.select_dtypes(exclude=['object'])
    X = obj_df.loc[:, obj_df.columns != 'Enjoy']
    obj_df['Enjoy'] = obj_df['Enjoy'].map({'Yes':1,'No':0})
    y = obj_df['Enjoy']
    return X, y
```

```python
def main():
    df = read_csv("../data/dt-data.csv")
    df = df.dropna()
    #obj_df = obj_df.fillna({"Occupied": "High"})
    X_train, y_train = encoding(df)
    df = read_csv("../data/test.csv")
    df = df.dropna()
    X_test, y_test = encoding(df)
    return
    
```

```python
df = read_csv("../data/test.csv")
df = df.dropna()
X_test, y_test = encoding(df)
y_test
```

```python
main()
model = tree.DecisionTreeClassifier()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
y_predict
list(y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```
