```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/dt-data.csv")
```

```python
df.columns = df.columns.str.replace('\W','')
df['Occupied'] = df['Occupied'].str.replace('[^[a-zA-Z]', '')
df['Enjoy'] = df['Enjoy'].str.replace('[^[a-zA-Z]', '')
```

```python
df.head()
df.dtypes
```

```python
obj_df = df.select_dtypes(include=['object']).copy()
```

```python
obj_df = obj_df.fillna({"Occupied": "High"})
```

```python
list(obj_df.columns)[:-1]
```

```python
X_features = pd.get_dummies(obj_df, columns=list(obj_df.columns)[:-1])
X_features.head()
```

```python
X_feature_encoded = X_features.select_dtypes(exclude=['object'])
X_feature_encoded.head()
```

```python
obj_df['Enjoy'] = obj_df["Enjoy"].astype('category')
obj_df['Enjoy'] = obj_df["Enjoy"].cat.codes
y_feature = obj_df['Enjoy']
y_feature.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_feature_encoded, y_feature, random_state=1)
```

```python
X_test
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```
